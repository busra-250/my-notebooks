### List comprehensions (üreteçler)
liste oluşturmaya yarar.kodları tek satır gibi yazmaya yarar.Kısa,okunabilir ve verimli bir şekilde liste oluşturmayı sağlayan bir yapıdır.Normalde bir listeyi oluşturmak için for döngüsüyle birkaç satır yazman gerekirken,list comprehension bunu tek satırda yapmanı sağlar.

```python
for item in list:
    if conditional:
        expression

[expression for item in list if conditional]
```

```python
liste=[]
for i in range(1000):
    liste+=[i]

print(liste)
```

```python
liste=[]
for i in range(1000):
    liste+=[i]

liste=[i for i in range(1000)]
print(liste)
```

```python
veri=["Ali",85,"Ayşe",92,"Mehmet",78,"Zeynep",90]
print("Orijinal liste:",veri)
ogrenci_dict={}
for i in range(0,len(veri),2):
    isim=veri[i]
    notu=veri[i+1]
    ogrenci_dict[isim]=notu
print(ogrenci_dict)

en_yuksek=max(ogrenci_dict.values())

for isim, notu in ogrenci_dict.items():
    if notu==en_yuksek:
        print("En yüksek notu alan öğrenci:",isim)

   
```
